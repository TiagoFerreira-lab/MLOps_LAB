```python
import plotly.graph_objects as go
import numpy as np

# Generate the spiral data
t = np.linspace(0, 4 * np.pi, 100)
x = np.sin(t)
y = np.cos(t)
z = t

# Create the 3D scatter plot with lines and markers
fig = go.Figure(data=[go.Scatter3d(
    x=x, 
    y=y, 
    z=z,
    mode='lines+markers',
    marker=dict(
        size=4,
        color=z,           # Color by the z-value for a gradient effect
        colorscale='Viridis'
    ),
    line=dict(
        color='magenta',   # The line color for the spiral
        width=3
    )
)])

# Update layout to add titles and axis labels
fig.update_layout(
    title='Cool 3D Spiral Plot',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)

# Display the interactive plot
fig.show()
```
